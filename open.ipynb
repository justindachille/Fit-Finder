```python
import sqlite3
import pandas as pd
from datetime import datetime

# Connect to the SQLite database
conn = sqlite3.connect('job_listings.db')

# Read the job listings table into a DataFrame
df = pd.read_sql_query("SELECT * FROM job_listings", conn)

# Close the database connection
conn.close()

# Display the DataFrame
display(df['link'][0])

# Calculate the number of job listings without sponsorship checked
sponsorship_unchecked = len(df[df['sponsorship_checked'] == 0])

# Calculate the number of job listings without candidate fit checked
candidate_fit_unchecked = len(df[df['candidate_fit_checked'] == 0])

# Find the oldest and newest dates of job listings
oldest_date = min(pd.to_datetime(df['published']))
newest_date = max(pd.to_datetime(df['published']))

# Display the results
print(f"Number of job listings without sponsorship checked: {sponsorship_unchecked}")
print(f"Number of job listings without candidate fit checked: {candidate_fit_unchecked}")
print(f"Oldest job listing date: {oldest_date}")
print(f"Newest job listing date: {newest_date}")
```
